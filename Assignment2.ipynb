```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the data
pinchez = pd.read_csv('/content/Pinchez/MyDrive/Colab Notebooks/Pinchez/new assignment2.csv')

# Extract the independent variable (X) and dependent variable (y)
X = pinchez[['Traffic_volume']]
y = pinchez['Accident_severity']

# Create and fit a linear regression model
regressor = LinearRegression()
regressor.fit(X, y)

# Plot the data points
pinchez.plot(x='Traffic_volume', y='Accident_severity', style='o', label='Data Points')

# Overlay the regression line
plt.plot(X, regressor.predict(X), color='red', linewidth=2, label='Regression Line')
plt.xlabel('Traffic_volume')
plt.ylabel('Accident_severity')
plt.legend()
plt.show()
```
